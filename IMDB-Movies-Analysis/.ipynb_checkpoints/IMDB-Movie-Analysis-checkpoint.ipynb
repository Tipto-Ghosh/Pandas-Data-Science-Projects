<h4> Percentance Of Missing Value in each columns

```python
# Function To Calculated Percentage

def get_missing_value_percentage(df , col_name: str) -> float: 
    total_rows = df.shape[0] 
    missing_count = df[col_name].isna().sum()
    percent = (missing_count / total_rows) * 100
    return percent if total_rows > 0 else 0

for column in movies.columns: 
    percent = get_missing_value_percentage(movies , column)
    print(f"Columns: {column} ---> missing values: {percent:.2f}%")
```

<h3> Convert Crew and genre Columns data in list format

```python
# Function to convert list to string 
def stringToList(data: str) -> list[str]: 
    # we know that there is no NaN value but to make it more general
    # Check value is not NaN if NaN return empty otherwise make a list of string 
    if isinstance(data , str):   
        data_list = data.split(',')
        # There can be leading space's or extra space at the end so remove that spaces
        data_list = [item.strip() for item in data_list]
        return data_list
    else:
        return []
```

```python
# converting crew and genre column from String to list     
movies['crew'] = movies['crew'].apply(stringToList)
movies['genre'] = movies['genre'].apply(stringToList)
```

<h3> Lead Actors Who's Movie Earned Most

```python
# Add a new column for lead actor (handle missing values)
movies['lead_actor'] = movies['crew'].apply(lambda crew_list: crew_list[0] if isinstance(crew_list, list) and crew_list else 'Missing Lead Actor')

# Group by lead actor and sum revenue
lead_actor_groupped_by_rev = movies.groupby(by='lead_actor', observed=True)['revenue'].sum()

# Reset index to make 'lead_actor' a normal column
most_revenued_lead_actors = lead_actor_groupped_by_rev.reset_index() 

# Rename column 
most_revenued_lead_actors.rename(columns={'lead_actor': 'actor_name'}, inplace=True) 

# Sort actors by revenue and get the top one
most_rev_earned_actor = most_revenued_lead_actors.sort_values(by='revenue', ascending=False).head(1)

# Extract values correctly
actor_name = most_rev_earned_actor.iloc[0]['actor_name']
total_revenue = most_rev_earned_actor.iloc[0]['revenue']

print(f"Most Revenue-Making Actor is: {actor_name}")
print(f"His/Her Total Revenue: {total_revenue}") 
```

<h3> Most Scored Lead Actor on IMDB

```python
# Groupping based on Lead Actor and also finding the total score a lead actor  
lead_actor_groupby_score = movies.groupby(by = 'lead_actor' , observed = True)['score'].sum()

# Reset index to convert Series to DataFrame
lead_actor_groupby_score = lead_actor_groupby_score.reset_index()

# Making lead_actor as index 
lead_actor_groupby_score.set_index('lead_actor' , inplace = True)
lead_actor_groupby_score.sort_values(by = 'score' , ascending = False , inplace = True) 

most_scored_actor = lead_actor_groupby_score.head(1)

most_scored_actor_name = most_scored_actor.index[0]
score = most_scored_actor['score'].iloc[0]
print(f"Most Scored Actor: {most_scored_actor_name}")
print(f"His/Her total Score: {score}")
```

<h3> Popular Genres

```python
# Group the movies based on the count of movies and avg scores
# movies.groupby(by = 'genre' , as_index = False).agg({'movie_name' : 'count' , 'score' : 'mean'}) 

# Movies genre is in list format so we have to split the genre in different rows 
movies_explode = movies.explode(column = 'genre')
# movies_explode.head() 
# Now group movies based on genre 
popular_genres = movies_explode.groupby(by = 'genre' , as_index = False).agg(movie_count = ('movie_name' , 'count') , 
                                                                             avg_score = ('score' , 'mean'))
```

```python
# plot the top 10 genre 
top10_genres = popular_genres.sort_values(by = ['movie_count' , 'avg_score'] , ascending = False).head(10) 

# Making the figure size 
plt.figure(figsize = (12 , 6)) 

# Creating Bar width and x positions 
x = np.arange(len(top10_genres))
bar_width = 0.4 

# plotting the bar 
plt.bar(x - bar_width / 2 , top10_genres['movie_count'] , width = bar_width , label = 'Movie Count' , color = 'royalblue')
plt.bar(x + bar_width / 2 , top10_genres['avg_score'] , width = bar_width , label = 'Avg. Score' , color = 'tomato')

# Adding labels and title
plt.xticks(ticks=x, labels=top10_genres['genre'], rotation=45, ha='right')  
plt.xlabel("Genre")
plt.ylabel("Count / Score")
plt.title("Top 10 Movie Genres: Movie Count vs. Average Score")
plt.legend()

plt.show() 
```

<h4> Take top 10 countries with most number of movies

```python
most_num_movies = movies['release_country'].value_counts()
most_num_movies = most_num_movies.reset_index()
most_num_movies.set_index('release_country' ,inplace = True)

top10_country_with_maximum_num_movies = most_num_movies.sort_values(by = 'count' ,ascending = False).head(10)
```

```python
top10_country_with_maximum_num_movies
```

<h3> Movie Industry Growth In Different Countries

```python
# Checking the growth of movie industry in top 10 most movie making countries 

# First filter all the movies that are made by top 10 countries 
movies_released_by_top10 = movies[movies['release_country'].isin(top10_country_with_maximum_num_movies.index)]

# Group movies bsed on Country and year 
movies_per_year = movies_released_by_top10.groupby(by = ['release_year' , 'release_country'] , 
                                                   as_index = False).agg(total_movies = ('movie_name' , 'count'))

# Create a pivot table 
pivot_movies = movies_per_year.pivot(index = 'release_year' , columns = 'release_country' , values = 'total_movies')

pivot_movies 
```

<h4> There are few years where some country make no movies. So replace NaN with zero

```python
pivot_movies.fillna(0 ,inplace = True)
```

```python
# Converting count movies into int32 

# pivot_movies[pivot_movies.columns] = pivot_movies[pivot_movies.columns].astype(int) 
pivot_movies = pivot_movies.astype(int)
pivot_movies 
```

<h4> Plotting Countries Movie Release Growth

```python
# Define figure and axes (5 rows, 2 columns)
fig, axes = plt.subplots(nrows=5, ncols=2, figsize=(15, 20))  # 10 subplots (5x2)

axes = axes.flatten()

# Iterate over the country columns
for idx, country in enumerate(pivot_movies.columns[1:11]):  # Skipping the 'release_year' column
    sns.lineplot(data=pivot_movies, x='release_year', y=country, ax=axes[idx])
    axes[idx].set_title(f"Industry Growth in {country}")
    axes[idx].set_xlabel("Year")
    axes[idx].set_ylabel("Total Movies") 

plt.tight_layout()
plt.show()
```
